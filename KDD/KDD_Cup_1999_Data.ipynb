### **공격타입**

- **Denial of Service (Dos):** Attacker tries to prevent
  legitimate users from using a service.
- **Remote to Local (r2l):** Attacker does not have an
  account on the victim machine, hence tries to gain
  access.
- **User to Root (u2r):** Attacker has local access to the
  victim machine and tries to gain super user privileges.
- **Probe:** Attacker tries to gain information about the
  target host.

![image](https://user-images.githubusercontent.com/42408554/65958381-35c16f80-e48a-11e9-80d6-8acc9fa8ff16.png)

---

- **root_shell:** '관리자 쉘' 보유 여부 확인 ***(U2R)***

## **1) 'root_shell'과 공격타입 'U2R' 관계 분석**
* '루트 계정'은 '최상위 권한자(슈퍼유저)'를 의미합니다. 따라서 '**root_shell**' 실행 권한을 가지고 있다는 것은 '루트 계정'을 보유하고 있다는 것으로 판단했습니다.
* 따라서 **root_shell** 컬럼 값이 0일 경우, **'U2R'**의 빈도가 높을 것이라고 생각했습니다. 그 이유는 '루트 계정'을 가지고 있어야만 루트 명령어 사용이 가능하기 때문에 '루트 계정'을 획득하는 과정이 필요하다고 판단했습니다.

* ***참고사이트***
    * root account: https://geek-university.com/linux/root-account/

### **공격타입 'Normal' 제외**

```python
root_shell_data = data[data['attack_type'] != 'Normal']
root_shell_data.head()
```

### **'root_shell'이 1인 값 제외**

```python
root_shell_data = root_shell_data[root_shell_data['root_shell'] != 1]
root_shell_data.head()
```

```python
root_shell_r2l = pd.DataFrame(root_shell_data.groupby(['attack_type'])['attack_type'].count())
root_shell_r2l = root_shell_r2l.rename(columns = {"attack_type": "attack_type_counts"})
root_shell_r2l = root_shell_r2l.sort_values(["attack_type_counts"], ascending=[False]).reset_index()
root_shell_r2l
```

```python
(ggplot(root_shell_r2l)
+ aes(x = 'attack_type', y = 'attack_type_counts', fill='attack_type')
+ geom_bar(stat = 'identity', position = 'dodge')
+ ggtitle('Relation between root_shell and U2R')
+ theme(axis_text_x = element_text(rotation = 60),
        figure_size = (8, 4))
)
```

### **결과**
- 'root_shell'과 'U2R'의 상관관계를 시각화한 결과, 초기 가정과 다르게 'root_shell'과 'U2R'은 큰 상관관계가 없다는 사실을 얻었습니다. 오히려 'Dos'의 비율이 월등히 높은 것으로 나타났습니다.

## **3) 'num_root', 'num_file_creations' 컬럼과 'U2R' 관계 분석**

- **'num_root'** 컬럼은 '루트 계정'에 접근한 횟수를 의미합니다. 만약 침입자가 '루트 계정'을 획득했을 경우, **'num_root'**의 횟수는 늘어날 것이고 해킹 또는 바이러스 침임을 위한 새로운 파일을 생성하여 **'num_file_creations'** 횟수도 늘어났을 것으로 판단했습니다.
- 또한, **'num_root'** 횟수이기 때문에 공격 타입은 **'U2R'**으로 판단했습니다.
- ***참고사이트***
    - Table 1- uploaded by Ralf C. Staudemeyer: https://www.researchgate.net/figure/The-41-features-provided-by-the-KDD-Cup-99-datasets_tbl1_263274883

```python
num_root = data[data['attack_type'] != 'Normal']
num_root = num_root[num_root['num_root'] >= 10].reset_index()
del num_root['index']
num_root
```

```python
num_root_file_creation = num_root[['num_root', 'num_file_creations', 'attack_type']]
num_root_file_creation
```

```python
fig = plt.figure(num=None, figsize=(10, 5), dpi=100, facecolor='w', edgecolor='k')
 
num_root = fig.add_subplot(2,2,1)
num_file_creations = fig.add_subplot(2,2,2)

num_root.set_title("num_root", fontsize=15)
num_file_creations.set_title("num_file_creations", fontsize=15)

num_root.plot(num_root_file_creation['num_root'],'k')
num_file_creations.plot(num_root_file_creation['num_file_creations'],'r-')
```

### **결과**
1. **'num_root'**와 **'num_file_creations'**가 비례하여 움직이지는 않지만, **'num_root'**의 횟수가 어느 정도 있어야만 **'num_file_creations'**의 횟수도 존재하는 것을 파악했습니다.
2. **'attack_type'**도 대부분 **'U2R'**인 점도 확인할 수 있었습니다.

## **4) 'count' 컬럼과 공격 타입 'Probe'의 관계 분석**

- **'Medium.com'**의 **'A Deeper Dive into the NSL-KDD Data Set' (링크 아래)**을 보면 공격 타입 **'Probe'**를 네트워크 상에서 고객의 정보나 은행 정보를 훔치는 것으로 설명하고 있습니다. 또한, **'count'** 컬럼의 설명을 보면 같은 호스트가 2초 이내로 현재 연결을 반복한 횟수를 나타냅니다.
- 따라서, 공격 타입이 **'Probe'**일 경우, **'count'** 컬럼값이 높을 가능성이 있다고 판단했습니다.
- ***참고사이트***
  - A Deeper Dive into the NSL-KDD Data Set: https://towardsdatascience.com/a-deeper-dive-into-the-nsl-kdd-data-set-15c753364657
  - Table 1- uploaded by Ralf C. Staudemeyer: https://www.researchgate.net/figure/The-41-features-provided-by-the-KDD-Cup-99-datasets_tbl1_263274883
  - [IT열쇳말] 서비스형 랜섬웨어(RaaS): http://www.bloter.net/archives/287539

```python
probe_data = data[data['attack_type'] == 'Probe']
probe_data = probe_data[probe_data['count'] >= 10].reset_index()
del probe_data['index']
probe_data.head()
```

```python
probe_data_connection_type = pd.DataFrame(probe_data.groupby(['attack_type', 'connection_type'])['connection_type'].count())
probe_data_connection_type = probe_data_connection_type.rename(columns = {"connection_type": "connection_type_counts"})
probe_data_connection_type = probe_data_connection_type.sort_values(["connection_type_counts"], ascending=[False]).reset_index()
probe_data_connection_type
```

```python
(ggplot(probe_data_connection_type)
+ aes(x = 'connection_type', y = 'connection_type_counts', fill = 'connection_type')
+ geom_bar(stat = 'identity', position = 'dodge')
+ ggtitle('Relation between Probe and Connection Type')
+ theme(axis_text_x = element_text(rotation = 60),
        figure_size = (8, 4))
)
```

```python
count_connection_type = pd.DataFrame(probe_data.groupby(['connection_type'])['count'].sum())
count_connection_type = count_connection_type.sort_values(["count"], ascending=[False]).reset_index()
count_connection_type
```

```python
(ggplot(count_connection_type)
+ aes(x = 'connection_type', y = 'count', fill = 'connection_type')
+ geom_bar(stat = 'identity', position = 'dodge')
+ ggtitle('Relation between Connection type and Count')
+ theme(text = element_text(family = 'BMJUA_ttf_0'),
       axis_text_x = element_text(rotation = 60),
       figure_size = (8, 4))
)
```

### **결과**
1. 공격 타입 **'Probe'**를 보면 **'count'** 컬럼값이 확실히 높은 것을 확인할 수 있었습니다. 그 이유는 위에서 언급한 대로 **'Probe'**은 사용자의 개인정보를 훔치는 공격이므로 **'count'** 값이 높다는 것은 매크로와 같이 인위적인 방법으로 사용자의 개인정보를 훔치려는 시도를 하는 것으로 파악했습니다.
2. 또한, **'connection_type'** 중에서 **'Satan'**이 가장 높은 **'count'** 값을 가지고 있는 것으로 나타났습니다. **'Satan'**은 돈 없이도 랜섬웨어 공격을 벌일 수 있는 RaaS 사례로서 개인정보를 훔치는 **'Probe'** 공격 타입에 가장 적절한 방식이라 판단했습니다.
